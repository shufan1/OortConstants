## 3.1 Simulation
With 500000 simulated test particles, we examined the effect of different sampling criteria. 


#### Talking about $\mu_l$ only: $\mu_l = (A\cos 2l+ B \sin 2l+C) \cos b$
Fig x1 shows the result of mu_l vs l function under three sample filtering criteria on galactic distance for test particles with |b|<20

$\mu_l \approx 0$ at 0 and 180, most negative at 90 and 270. Larger dispersion at 90 and 270 while lowest at 0 and 180. 

Discussion: at $l=0 / 180$, if a star is in circular orbit with radius $R$, the horizontal compoent of the star's tangential velocity with respect to the Sun is $v_{circ}(R)-v_{circ}(R_\odot)$. And when $R$ is close to $R_\odot$, this difference is close to 0.

at $l=90 / 270$, this tangential component of relative velocity is the most negative because ...
